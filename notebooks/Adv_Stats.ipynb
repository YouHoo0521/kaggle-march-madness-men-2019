We notice that each box score contains two teams: a winning team and a losing team. Thus, each row corresponds to one game. We would like give each team its own row, such that there are two rows per game. This will allow us to calculate advanced stats for each team.

```python
def setWinAndLoseTeamsRecords(RegularSeasonDetailedResults):
    #Convert the data frame from one record per game to one record per team/game
    regSeaW = RegularSeasonDetailedResults.rename(columns = {'WTeamID': 'TEAMID',
                                                             'WScore': 'SCORE',
                                                             'WFGM': 'FGM',
                                                             'WFGA': 'FGA',
                                                             'WFGM3': 'FGM3',
                                                             'WFGA3': 'FGA3',
                                                             'WFTM': 'FTM',
                                                             'WFTA': 'FTA',
                                                             'WOR': 'OR',
                                                             'WDR': 'DR',
                                                             'WAst': 'AST',
                                                             'WTO': 'TO',
                                                             'WStl': 'STL',
                                                             'WBlk': 'BLK',
                                                             'WPF': 'PF',
                                                             'LTeamID': 'O_TEAMID',
                                                             'LScore': 'O_SCORE',
                                                             'LFGM': 'O_FGM',
                                                             'LFGA': 'O_FGA',
                                                             'LFGM3': 'O_FGM3',
                                                             'LFGA3': 'O_FGA3',
                                                             'LFTM': 'O_FTM',
                                                             'LFTA': 'O_FTA',
                                                             'LOR': 'O_OR',
                                                             'LDR': 'O_DR',
                                                             'LAst': 'O_AST',
                                                             'LTO': 'O_TO',
                                                             'LStl': 'O_STL',
                                                             'LBlk': 'O_BLK',
                                                             'LPF': 'O_PF'
                                                            })

    regSeaL = RegularSeasonDetailedResults.rename(columns = {'LTeamID': 'TEAMID',
                                                             'LScore': 'SCORE',
                                                             'LFGM': 'FGM',
                                                             'LFGA': 'FGA',
                                                             'LFGM3': 'FGM3',
                                                             'LFGA3': 'FGA3',
                                                             'LFTM': 'FTM',
                                                             'LFTA': 'FTA',
                                                             'LOR': 'OR',
                                                             'LDR': 'DR',
                                                             'LAst': 'AST',
                                                             'LTO': 'TO',
                                                             'LStl': 'STL',
                                                             'LBlk': 'BLK',
                                                             'LPF': 'PF',

                                                             'WTeamID': 'O_TEAMID',
                                                             'WScore': 'O_SCORE',
                                                             'WFGM': 'O_FGM',
                                                             'WFGA': 'O_FGA',
                                                             'WFGM3': 'O_FGM3',
                                                             'WFGA3': 'O_FGA3',
                                                             'WFTM': 'O_FTM',
                                                             'WFTA': 'O_FTA',
                                                             'WOR': 'O_OR',
                                                             'WDR': 'O_DR',
                                                             'WAst': 'O_AST',
                                                             'WTO': 'O_TO',
                                                             'WStl': 'O_STL',
                                                             'WBlk': 'O_BLK',
                                                             'WPF': 'O_PF',
                                                             })

    regSea = (regSeaW, regSeaL)
    regSea = pd.concat(regSea, ignore_index = True, sort = False)
    regSea = regSea[['Season','TEAMID', 'DayNum', 'SCORE', 'O_TEAMID', 'O_SCORE',
                 'FGM', 'FGA', 'FGM3', 'FGA3', 'FTM', 'FTA', 'OR', 'DR',
                 'AST', 'TO', 'STL', 'BLK', 'PF', 'NumOT',
                 'O_FGM', 'O_FGA', 'O_FGM3', 'O_FGA3', 'O_FTM', 'O_FTA', 'O_OR', 'O_DR',
                 'O_AST', 'O_TO', 'O_STL', 'O_BLK', 'O_PF', 'GameNum'
                 ]]
    return regSea
```

```python
RegularSeasonDetailedResults['GameNum'] = 1
regSes = setWinAndLoseTeamsRecords(RegularSeasonDetailedResults)
regSes.head()
```

```python
def GetAdvancedStats(Game_Min,regsea_box):
    # effective field goal %
    regsea_box['EFG']       = (regsea_box['FGM'] + (regsea_box['FGM3']*0.5))/regsea_box['FGA']
    
    # turnover rate
    regsea_box['TOV']       = regsea_box['TO']/((regsea_box['FGA'] + 0.44) + (regsea_box['FTA']+regsea_box['TO']))
    
    # off and def rebound ratios
    regsea_box['ORB']       = regsea_box['OR']/(regsea_box['OR'] + regsea_box['O_DR'])
    regsea_box['DRB']       = regsea_box['DR']/(regsea_box['DR'] + regsea_box['O_OR'])
    
    # free throw attempt ratio
    regsea_box['FTAR']      = regsea_box['FTA']/(regsea_box['FGA'])
    regsea_box['O_FTAR']    = regsea_box['O_FTA']/(regsea_box['O_FGA'])
    
    # true shooting %
    regsea_box['TS']        = regsea_box['SCORE']/((regsea_box['FGA']*2) + (0.88 * regsea_box['FTA']))
    regsea_box['O_TS']      =  regsea_box['O_SCORE']/((regsea_box['O_FGA']*2) + (0.88 * regsea_box['O_FTA']))
    
    # assist to turnover ratio
    regsea_box['ASTTO']     = (regsea_box['AST']/(regsea_box['TO']))
    regsea_box['O_ASTTO']   = (regsea_box['O_AST']/(regsea_box['O_TO']))
    
    # assist ratio
    regsea_box['ASTR']      = (regsea_box['AST'] * 100) / ( (regsea_box['FGA'] + (regsea_box['FTA']*0.44)) + regsea_box['AST'] + regsea_box['TO'] )
    regsea_box['O_ASTR']    = (regsea_box['O_AST'] * 100) / ( (regsea_box['O_FGA'] + (regsea_box['O_FTA']*0.44)) + regsea_box['O_AST'] + regsea_box['O_TO'] )

    # total rebounds (team and opponent) and rebound percentage
    regsea_box['TR']        = regsea_box['OR'] + regsea_box['DR']
    regsea_box['O_TR']      = regsea_box['O_OR'] + regsea_box['O_DR']
    regsea_box['REBP']      = 100 * (regsea_box['TR']) / (regsea_box['TR'] + regsea_box['O_TR'])
    
    # number of possessions
    regsea_box['POSS']      = 0.5 * ((regsea_box['FGA'] + 0.4 * regsea_box['FTA'] - 1.07 * (regsea_box['OR'] / (regsea_box['OR'] + regsea_box['O_DR'])) * (regsea_box['FGA'] - regsea_box['FGM']) + regsea_box['TO']) + (regsea_box['O_FGA'] + 0.4 * regsea_box['O_FTA'] - 1.07 * (regsea_box['O_OR'] / (regsea_box['O_OR'] + regsea_box['DR'])) * (regsea_box['O_FGA'] - regsea_box['O_FGM']) + regsea_box['O_TO']))
    regsea_box['O_POSS']    = 0.5 * ((regsea_box['O_FGA'] + 0.4 * regsea_box['O_FTA'] - 1.07 * (regsea_box['O_OR'] / (regsea_box['O_OR'] + regsea_box['DR'])) * (regsea_box['O_FGA'] - regsea_box['O_FGM']) + regsea_box['O_TO']) + (regsea_box['FGA'] + 0.4 * regsea_box['FTA'] - 1.07 * (regsea_box['OR'] / (regsea_box['OR'] + regsea_box['O_DR'])) * (regsea_box['FGA'] - regsea_box['FGM']) + regsea_box['TO']))
    
    # game minutes
    regsea_box['GM']        = (Game_Min*regsea_box['GameNum']) + (5*regsea_box['NumOT'])
    
    # pace
    regsea_box['PACE']      = 40 * ((regsea_box['POSS'] ) / (2 * (regsea_box['GM'] / 5)))
    
    # def and off rating
    regsea_box['DRTG']      = 100* (regsea_box['O_SCORE']/regsea_box['POSS'])
    regsea_box['ORTG']      = 100* (regsea_box['SCORE']/(regsea_box['O_POSS']))
    
    # 3 fg% and defending the 3
    regsea_box['OFF3']      = (regsea_box['FGM3']/(regsea_box['FGA3']))
    regsea_box['DEF3']      = (regsea_box['O_FGM3']/(regsea_box['O_FGA3']))
    
    # opponent effective field goal %
    regsea_box['O_EFG']     = (regsea_box['O_FGM'] + (regsea_box['O_FGM3']*0.5))/regsea_box['O_FGA']
    
    # opponent turnover rate
    regsea_box['O_TOV']     = regsea_box['O_TO']/((regsea_box['O_FGA'] + 0.44) + (regsea_box['O_FTA']+regsea_box['O_TO']))
    
    # another def and off rating (hmmm)
    regsea_box['DEFRTG']    = 100*regsea_box['O_SCORE']/(regsea_box['O_FGA'] + regsea_box['O_TO'] + (0.44* regsea_box['O_FTA']) - regsea_box['O_OR'])
    regsea_box['OFFRTG']    = 100*regsea_box['SCORE']/(regsea_box['FGA'] + regsea_box['TO'] + (0.44*regsea_box['FTA']) - regsea_box['OR'])
    
    # another turnover rate? (hmmm)
    regsea_box['TOR']       = (regsea_box['TO'] * 100) / (regsea_box['FGA'] + (regsea_box['FTA'] * 0.44) + regsea_box['AST'] + regsea_box['TO'])
    regsea_box['O_TOR']     = (regsea_box['O_TO'] * 100) / (regsea_box['O_FGA'] + (regsea_box['O_FTA'] * 0.44) + regsea_box['O_AST'] + regsea_box['O_TO'])
    
    # steal to turnover ratio
    regsea_box['STLTO']     = regsea_box['STL']/regsea_box['TO']
    regsea_box['O_STLTO']   = regsea_box['O_STL']/regsea_box['O_TO']
    
    # performance impact estimator
    regsea_box['PIE']       = (regsea_box['SCORE'] + regsea_box['FGM'] + regsea_box['FTM'] - regsea_box['FGA']  - regsea_box['FTA']  + regsea_box['DR'] + (.5 * regsea_box['OR']) + regsea_box['AST'] + regsea_box['STL'] + (.5 * regsea_box['BLK']) - regsea_box['PF'] - regsea_box['TO']) / ((regsea_box['SCORE'] + regsea_box['FGM'] + regsea_box['FTM'] - regsea_box['FGA']  - regsea_box['FTA']  + regsea_box['DR'] + (.5 * regsea_box['OR']) + regsea_box['AST'] + regsea_box['STL'] + (.5 * regsea_box['BLK']) - regsea_box['PF'] - regsea_box['TO'])  + (regsea_box['O_SCORE'] + regsea_box['O_FGM'] + regsea_box['O_FTM'] - regsea_box['O_FGA'] - regsea_box['O_FTA'] + regsea_box['O_DR'] + (.5 * regsea_box['O_OR']) + regsea_box['O_AST'] + regsea_box['O_STL'] + (.5 * regsea_box['O_BLK']) - regsea_box['O_PF'] - regsea_box['O_TO']))
        
    return regsea_box
```
