# Boxscore Dataset

The boxscore dataset describes the performance of each team in the regular season.

- Each row represents a game between team1 (lower id) and team2 (higher id)
- For each boxscore statistic (e.g. ast - assists),
    - ast_team_mean1: average assists by team1
    - ast_team_std1: standard deviation of assists by team1
    - ast_opp_mean1: average assists by opponents of team1
    - ast_opp_std1: standard deviation of assists by opponents of team1
